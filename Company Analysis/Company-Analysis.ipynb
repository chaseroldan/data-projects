```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('companies.csv')
df.head()
```

```python
df2 = df.replace('billion','00000000', regex=True)
df3 = df2.replace('million','000000', regex=True)
df3[['Revenue 2018', 'Revenue 2019', 'Revenue 2020', 'Net Income 2018', 'Net Income 2019', 'Net Income 2020']] = df3[['Revenue 2018', 'Revenue 2019', 'Revenue 2020', 'Net Income 2018', 'Net Income 2019', 'Net Income 2020']].replace(' ','', regex=True)
df3[['Revenue 2018', 'Revenue 2019', 'Revenue 2020', 'Net Income 2018', 'Net Income 2019', 'Net Income 2020']] = df3[['Revenue 2018', 'Revenue 2019', 'Revenue 2020', 'Net Income 2018', 'Net Income 2019', 'Net Income 2020']].replace('\.', '', regex=True)
df3[['Revenue 2018', 'Revenue 2019', 'Revenue 2020', 'Net Income 2018', 'Net Income 2019', 'Net Income 2020']] = df3[['Revenue 2018', 'Revenue 2019', 'Revenue 2020', 'Net Income 2018', 'Net Income 2019', 'Net Income 2020']].replace('\$', '', regex=True)
df3[['Revenue 2018', 'Revenue 2019', 'Revenue 2020', 'Net Income 2018', 'Net Income 2019', 'Net Income 2020']] = df3[['Revenue 2018', 'Revenue 2019', 'Revenue 2020', 'Net Income 2018', 'Net Income 2019', 'Net Income 2020']].astype(float)
df3["Total Revenue 2018 - 2020"] = df3["Revenue 2018"] + df3["Revenue 2019"] + df3["Revenue 2020"]
df3["Total Net Income 2018 - 2020"] = df3["Net Income 2018"] + df3["Net Income 2019"] + df3["Net Income 2020"]
df3['Avg Revenue 2018 - 2020'] = df3['Total Revenue 2018 - 2020'] / 3
df3['Avg Net Income 2018 - 2020'] = df3['Total Net Income 2018 - 2020'] / 3
companies = df3
companies['Avg Profit Margin 2018 - 2020'] = companies['Avg Net Income 2018 - 2020'] / companies['Avg Revenue 2018 - 2020']

companies
```

```python
df1 = companies[['Company Name', 'Country Founded', 'Revenue 2018', 'Revenue 2019', 'Revenue 2020', 'Net Income 2018', 'Net Income 2019', 'Net Income 2020', 'Total Revenue 2018 - 2020', 'Avg Revenue 2018 - 2020', 'Total Net Income 2018 - 2020', 'Avg Net Income 2018 - 2020', 'Avg Profit Margin 2018 - 2020']]
df1
```

```python
duplicates = df1.groupby('Company Name', as_index=False).sum()
duplicates = duplicates.set_index('Company Name')
duplicates
```

```python
df2 = companies[['Company Name', 'Country Founded', 'Industry']].sort_values(by='Company Name', ascending=True).copy()
df2
```

```python
duplicates2 = df2.groupby('Company Name', as_index=False).sum()
duplicates2 = duplicates2.set_index('Company Name')
duplicates2
```

```python
merged = pd.merge(duplicates2, duplicates, left_index=True, right_index=True)
merged
```

```python
merged.to_csv('companies.csv')
```
