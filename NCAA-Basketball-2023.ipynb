```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read data csv

data = pd.read_csv('2023data.csv')

# fix column name
data.rename(columns = {'3:00 PM':'3PM'}, inplace = True)

# set index to team name
data = data.set_index('TEAM')

# view
data
```

```python
# read records csv
rec = pd.read_csv('2023records.csv')

# set index to team name
rec = rec.set_index('TEAM')

# view
rec
```

```python
# read csv
allowed = pd.read_csv('2023allowed.csv')

# set index to team
allowed = allowed.set_index('TEAM')

# view
allowed
```

```python
# merge data
combined = pd.merge(rec, data, left_index=True, right_index=True)
merged = pd.merge(combined, allowed, left_index=True, right_index=True)

# split win/loss column
merged[['W','L']] = merged['RECORD'].str.split('-',expand=True)

# make wins and losses numeric
merged["W"] = pd.to_numeric(merged["W"])
merged["L"] = pd.to_numeric(merged["L"])

# view
merged
```

```python
# calculate and add additional columns
merged["PTS-Ratio"] = merged["PTS"] / merged["PTS-ALLOWED"]
merged["Made"] = merged["FGM"] + merged["3PM"] + merged["FTM"]
merged["Attempts"] = merged["FGA"] + merged["3PA"] + merged["FTA"]
merged["Efficiency"] = merged["PTS"] / merged["Made"]
merged["Accuracy"] = merged["Made"] / merged["Attempts"]
merged.head()
```

```python
# create new csv
merged.to_csv('2023-complete-data.csv')
```
